```python
import pandas as pd
import random

file_path = 'dataset.csv'
data = pd.read_csv(file_path)

# Membuat 500 data acak baru
random_texts = [f"Teks random {i}" for i in range(1, 501)]
random_labels = [random.choice(data['label'].unique()) for _ in range(500)]

# Membuat DataFrame baru dari data acak
new_data = pd.DataFrame({'text': random_texts, 'label': random_labels})

# Menggabungkan dengan data asli
data_extended = pd.concat([data, new_data], ignore_index=True)

# Menampilkan 5 data terakhir sebagai contoh
# print(data_extended.tail())
data_extended.tail()
```
